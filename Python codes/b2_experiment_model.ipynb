```python
# =========================
# Library imports
# =========================

import numpy as np # Numerical operations and matrix manipulation
import matplotlib.pyplot as plt # Plotting and visualization library


# =========================
# Figure formatting setup
# =========================

# Configures default plot appearance
plt.rcParams.update({

    # Uses serif fonts in all figures
    "font.family": "serif",

    # Sets Times New Roman as the main font
    "font.serif": ["Times New Roman"],

    # Uses STIX fonts for mathematical expressions
    "mathtext.fontset": "stix",

    # Ensures editable text in exported PDF files
    "pdf.fonttype": 42
})
```

```python
# Load data from text file
data = np.loadtxt("b2Experiment.txt")

# Extract and shift time and velocity data
# Note: Python uses 0-based indexing (column 1 is index 0)
t_data = data[:, 0] - 21.89
v_data = data[:, 1]

# Plot measured data
plt.figure(figsize=(8, 6))
plt.plot(t_data, v_data, '-b', linewidth=1.1, label="measured data")

# Set labels and grid
plt.xlabel('Time (s)', fontsize=14, fontname='Times New Roman')
plt.ylabel(r'$\omega$ (rad/s)', fontsize=14, fontname='Times New Roman')
plt.grid(True)

# Define model parameters
t_model = np.linspace(0, 20, 1000)
w0 = 45.81   # Initial value
B = 2.1e-05
J = 9.6264e-05

# Calculate theoretical model
w_model = w0 * np.exp(-(B / J) * t_model)

# Plot model
plt.plot(t_model, w_model, 'r', linewidth=1.2, label="model")

# Final formatting
plt.legend(loc='best', fontsize=14)
plt.xlim([-0.5, 7.2])
plt.ylim([-2, 50])

# Saves figure
plt.savefig('b2Experiment.pdf',
            format='pdf',
            bbox_inches='tight',
            facecolor='white')

# Show the plot
plt.show()
```
